## Deliverable 2: Scrape and Analyze Mars Weather Data

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt

```

```python
browser = Browser('chrome')

```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)

```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
mars_soup = soup(html, 'html.parser')

```

```python
# Extract all rows of data
mars_rows = mars_soup.find_all(class_ = 'data-row')

```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
row_data =[]


# Loop through the scraped data to create a list of rows
for row in mars_rows:
    td_all = row.find_all('td')
    results = [td.text for td in td_all]
    row_data.append(results)
row_data


```

```python
# get the column header data
mars_header_all = mars_soup.find_all('th')

# extract the column header data into a list
mars_headers = [header.text for header in mars_header_all]
mars_headers

```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names

mars_df = pd.DataFrame(row_data)
mars_df.columns = mars_headers

```

```python
# Confirm DataFrame was created successfully
mars_df

```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_df.info()

```

```python
# Change data types for data analysis

# change "id", "sol", "ls", and "month" to int, change "min_temp", "pressure" to float
mars_df = mars_df.astype({"id":"int", "sol":"int", "ls":"int",
                     "month":"int", "min_temp":"float", "pressure":"float"})

# change "terrestrial_date" to datetime
mars_df['terrestrial_date'] =  pd.to_datetime(mars_df['terrestrial_date'])

```

```python
# Confirm type changes were successful by examining data types again
mars_df.info()

```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
months_on_mars = mars_df.month.nunique()
months_on_mars

print(f"Mars has {months_on_mars} months.")
```

```python
# 2. How many Martian days' worth of data are there?
martian_days_data = mars_df.sol.max()
martian_days_data

print(f"There are {martian_days_data} Martian days' worth of data.")

```

```python
# 3. What is the average low temperature by month?

mars_monthly_low_temp_avg = mars_df.groupby(["month"]).mean()["min_temp"]
mars_monthly_low_temp_df = mars_monthly_low_temp_avg.to_frame()
mars_monthly_low_temp_df = mars_monthly_low_temp_df.reset_index()
mars_monthly_low_temp_df.rename(columns = {'min_temp':'avg_min_temp'}, inplace = True)

mars_monthly_low_temp_df
```

```python
# Plot the average temperature by month

# Set the data for the axes
x_axis = mars_monthly_low_temp_df["month"]
y_axis = mars_monthly_low_temp_df["avg_min_temp"]

# Add labels
plt.title("Average Minimum \n Temp by Month on Mars")
plt.xlabel("Month")
plt.ylabel("Temp")

# Bar Plot
plt.bar(x_axis, y_axis, color = 'goldenrod')

# Display plot
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location

# find the temp for the coldest and hottest months

min_temp_mars = mars_monthly_low_temp_df['avg_min_temp'].min()
max_temp_mars = mars_monthly_low_temp_df['avg_min_temp'].max()

# find the corresponding month for the min and max temps found above

mars_coldest_month = mars_monthly_low_temp_df[mars_monthly_low_temp_df['avg_min_temp']==min_temp_mars]['month'].item()
mars_hottest_month = mars_monthly_low_temp_df[mars_monthly_low_temp_df['avg_min_temp']==max_temp_mars]['month'].item()

print(f"The month of {mars_coldest_month} with a temp of {min_temp_mars:.4f} is the coldest Martian month \n and the month of {mars_hottest_month} with a temp of {max_temp_mars:.4f} is the hottest Martian month.")

```

```python
# 4. Average pressure by Martian month
mars_monthly_pressure_avg = mars_df.groupby(["month"]).mean()["pressure"]

mars_monthly_pressure_avg_df = mars_monthly_pressure_avg.to_frame()
mars_monthly_pressure_avg_df = mars_monthly_pressure_avg_df.reset_index()
mars_monthly_pressure_avg_df.rename(columns = {'pressure':'avg_pressure'}, inplace = True)

mars_monthly_pressure_avg_df


```

```python
# Plot the average pressure by month

# Set the data for the axes
x_axis = mars_monthly_pressure_avg_df["month"]
y_axis = mars_monthly_pressure_avg_df["avg_pressure"]

# Add labels
plt.title("Average Pressure \n by Month on Mars")
plt.xlabel("Month")
plt.ylabel("Pressure")

# Bar Plot
plt.bar(x_axis, y_axis, color = 'plum')

# Display plot
plt.show()
```

```python
# Identify the lowest and highest pressure months on Mars

# find the pressure of the lowest and highest pressure months

min_pressure_mars = mars_monthly_pressure_avg_df['avg_pressure'].min()
max_pressure_mars = mars_monthly_pressure_avg_df['avg_pressure'].max()

# find the corresponding month for the min and max temps found above

mars_lowest_pressure_month = mars_monthly_pressure_avg_df[mars_monthly_pressure_avg_df['avg_pressure']==min_pressure_mars]['month'].item()
mars_highest_pressure_month = mars_monthly_pressure_avg_df[mars_monthly_pressure_avg_df['avg_pressure']==max_pressure_mars]['month'].item()

print(f"The month of {mars_lowest_pressure_month} with an average pressure of {min_pressure_mars:.4f} is the lowest pressure Martian month \n and the month of {mars_highest_pressure_month} with a pressure of {max_pressure_mars:.4f} is the highest pressure Martian month.")

```

## How many terrestrial (earth) days are there in a Martian year?
## Calculated in two ways: (1) visually (2) using the data in the table

```python
# Visual calculation of number of earth days in a Martian year

# Plot the min_temp vs terrestrial date

# Set the data for the axes
x_axis = mars_df["terrestrial_date"]
y_axis = mars_df["min_temp"]

# Add labels
plt.title("Minimum Temp vs \n Earth Days")
plt.xlabel("Earth Days")
plt.ylabel("Temp")

# Bar Plot
plt.bar(x_axis, y_axis, color = 'seagreen')

# Display plot
plt.show()
```

```python
# Appears to be low temp peaks in both 2014 and 2016
# thus, get those dates that have the lowest temp in each year

#-----------------------------------------------------------------
# make a dataframe for only those dates for 2014

df_2014 = mars_df.loc[(mars_df['terrestrial_date'] >= '2014-01-01')
                     & (mars_df['terrestrial_date'] <= '2014-12-31')]

# calculate the lowest temp within 2014

lowest_temp_2014 = df_2014['min_temp'].min()

# get the corresponding dates that have the lowest temp in 2014
lowest_temp_dates_2014_df = df_2014[df_2014['min_temp'] == lowest_temp_2014] 
lowest_temp_dates_2014_df
```

```python
# make a dataframe for only those dates for 2016

df_2016 = mars_df.loc[(mars_df['terrestrial_date'] >= '2016-01-01')
                     & (mars_df['terrestrial_date'] <= '2016-12-31')]

# calculate the lowest temp within 2016

lowest_temp_2016 = df_2016['min_temp'].min()

# get the corresponding dates that have the lowest temp in 2016
lowest_temp_dates_2016_df = df_2016[df_2016['min_temp'] == lowest_temp_2016] 
lowest_temp_dates_2016_df
```

```python
# calculate the difference between the lowest temp dates found for 2014 and 2016
# this calculation is the number of earth days in a martian year 
# as calculated visually looking at a chart of the terrestrial date plotted against min_temp

low_temp_2014 = dt.datetime(2014, 2, 3)
low_temp_2016 = dt.datetime(2016, 1, 31)


visual_martian_year = (low_temp_2016-low_temp_2014).days

print(f"By plotting minimum temp vs earth days and calculating the number of days \n between two minimum temp days, a Martian year is calculated as {visual_martian_year} days.")

```

```python
# Using table data to calculate number of earth days in a martian year

#--------------------------------------------------------

# Create a function to calculate the Martian year
def martian_year(num):
    
    # Isolate data from the 'ls' column where the values range from 0 through 359, set variable as 'num'
    solar_longitude_df = mars_df.loc[mars_df['ls'] == num]
    
    # sort the values of the filtered dataframe by 'terrestrial date'
    sorted_solar_longitude_df = solar_longitude_df.sort_values(by='terrestrial_date')
    
    # calculate the difference for each consecutive "terrestrial date" in sorted dataframe
    diff_df = sorted_solar_longitude_df.diff()
    
    # change the datatype to eliminate all data except for number of days
    diff_df['terrestrial_date'] = diff_df['terrestrial_date'].dt.days
    
    # calculate the max difference in "terrestrial date" as this will be the martian year
    martian_year = diff_df['terrestrial_date'].max()
    
    # return the value of the number of earth days in a martian year
    return martian_year

#--------------------------------------------------------
#--------------------------------------------------------


# iterate through all the values from 0 to 359 to get a list of days for a Martian year

# first, establish an empty list that will eventually contain all calculations of days in a Martian year
earth_days_list = []

# create a for loop to analyze all 359 data points of the solar longitude (ls) using the function above
for x in range(360):
    results = martian_year(x)
    earth_days_list.append(results)

# create a dataframe from the list created
ls_data = pd.DataFrame(earth_days_list)

# drop and nulls from the dataframe
ls_data_02 = ls_data.dropna()

# remove any outlier data
ls_data_df = ls_data_02.loc[(ls_data_02[0] >= 500) & (ls_data_02[0] <= 800)]

# Calculate the average of all the data, this is the number of terrestrial days in a Martian year
martian_year = ls_data_df[0].mean()
martian_year_std = ls_data_df[0].std()


print(f"There are {martian_year:.4f} terrestrial days in a Martian year \n with a standard deviation of {martian_year_std:.4f}.")

```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
